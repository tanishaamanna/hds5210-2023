## #04 -

Which of those State values has the best average performance on the `Star Rating for health team communicated well with them` score?

```python
# Write some code to figure it out
avg_scores_by_state = hhcaps.groupby('State')['Star Rating for health team communicated well with them'].mean()

# highest average score with state
best_state = avg_scores_by_state.idxmax()

# highest average score
highest_average_score = avg_scores_by_state.max()
```
