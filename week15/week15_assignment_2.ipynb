### 2. Query some data

Assuming the source database has some data in it or that you can load some data into it from any source, query it.  Then, read that data into a Pandas dataframe.

```python
import pandas as pd
```

```python
project_id = 'hds5210-407318'
dataset_id = "bigquery-public-data.covid19_weathersource_com"
table_id = "bigquery-public-data.covid19_weathersource_com.county_day_history"
```

```python
from google.cloud import bigquery

client = bigquery.Client(project=project_id)

query = f"""
    SELECT *
    FROM {'bigquery-public-data.covid19_weathersource_com.county_day_history'}
"""
query_job = client.query(query)
results = query_job.result()
covid_weathersource = results.to_dataframe()
print(covid_weathersource)
```

### 3. Aggregate your data frame

Do some kind of aggregation on your data frame.  Something that makes sense and has some groups to it.  Don't just sum up one column for the entire data frame.  Be more creative than that.

```python
query = f"""
    SELECT
        country,
        AVG(avg_temperature_air_2m_f) AS avg_temperature,
        AVG(avg_humidity_relative_2m_pct) AS avg_humidity
    FROM
       {'bigquery-public-data.covid19_weathersource_com.county_day_history'}
    WHERE
        avg_temperature_air_2m_f IS NOT NULL
        AND avg_humidity_relative_2m_pct IS NOT NULL
    GROUP BY
        country
"""
query_job = client.query(query)
results = query_job.result()
covid_weathersource= results.to_dataframe()
covid_weathersource
```

### 4. Write to Excel

Use Pandas functions to write your summarized data out to a local Excel file.

```python
excel_file_path = 'summarized_data.xlsx'
covid_weathersource.to_excel('summarized_weatherdata.xlsx', index=False)
print(f'File saved successfully: {excel_file_path}')
```
